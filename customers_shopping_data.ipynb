```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
df = pd.read_csv('customers_shopping.csv')
```

```python
#create the new column for age_group

labels = ['young adult' , 'adult' , 'middle aged' , 'senior']
df['age_group'] = pd.qcut(df['age'], q=4 , labels = labels)
```

```python
df[['age' , 'age_group']].head(10)
```

```python
labels = ['short' , 'medium' , 'long']
df['last_login_period'] = pd.qcut(df['last_login_days_ago'], q=3 , labels = labels)
```

```python
df[['last_login_period' , 'last_login_days_ago']].head(10)
```

```python
#Identifies high-risk age segment.

plt.figure(figsize=(6,4))
sns.countplot(data=df, x='age_group', hue='churn')
plt.title("Churn by Age Group")
plt.show()
```

```python
#Shows which plan generates most money.

revenue_by_sub = df.groupby('subscription_type')['total_charges'].sum()

plt.figure(figsize=(6,3))
sns.barplot(x=revenue_by_sub.index, y=revenue_by_sub.values)
plt.title("Revenue by Subscription Type")
plt.ylabel("Total Revenue")
plt.show()
```
